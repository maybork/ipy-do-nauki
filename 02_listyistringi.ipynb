Listy, tuples i str to przedstawiciele klasy 'iterowalnych'. Jak nazwa wskazuje, możemy po nich iterować. Robi się to z pomocą `for` (działa jak foreach w niektórych innych językach).

```python
lista = [1, 2, 3, 4, 5, 6]
print(lista)
```

```python
for liczba in lista:
    print(liczba)
```

```python
tekst = "kocham cię\n"
for litera in tekst:
    print(litera.upper())
```

Możemy uzyskać dostęp do konkretnego elementu listy używając jego indeksu, również z możliwością zmiany.

```python
lista = ["J", "P", "2"]
print(lista[2], lista)
```

```python
lista[2] = "3"
print(lista[2], lista)
```

Długość iterowalnego obiektu sprawdzamy funkcją `len()`.

```python
print(len(lista))
```

Listy są jednym z najbardziej elastycznych elementów języka. Bez problemu możemy wykonywać podstawowe operacje, nie martwiąc się własną implementacją (zwykle).

```python
usuwanie = [4, 2, 1]
usuwanie.remove(4)
print(usuwanie)
```

```python
dodawanie = [3]
dodawanie.append(7)
print(dodawanie)
```

```python
usuwanie.extend(dodawanie)
print(usuwanie)
```

```python
lista = ["a", "b"]
lista.reverse()
print(lista)
```
