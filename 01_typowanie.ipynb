Python jest językiem implicit typów.

```python
tekst = "Hejka"
print(type(tekst))
# Ten sam efekt możemy osiągnąć korzystając z konstruktora.
tekst = str("Hejka")
print(type(tekst))
```

Typ `str` ma wiele ciekawych metod.

```python
print(tekst.upper())
print(tekst.replace("H", "Nakl"))
```

Co ciekawe, nie zmienia wartości oryginalnej zmiennej. Jeśli taki jest twój cel, powinnaś przypisać wynik metody do zmiennej.

```python
print(tekst)
tekst = tekst.upper()
print(tekst)
```

Typowanie w Pythonie odbywa się dynamicznie - zwracamy uwagę nie na sam typ, a jego własności. Użycie `.upper()` na jakimś obiekcie jest okej, skoro została zaimplementowana!

```python
calkowita = 1
print(type(calkowita))
floatingpoint = 1.0
print(type(floatingpoint))
zlozona = 2 + 1j
print(type(zlozona))
```

Wyróżniamy parę operatorów arytmetycznych, są podobne do większości innych języków.

```python

# %% Wyróżniamy parę operatorów arytmetycznych
suma = 3 + 2
różnica = 3 - 2
iloczyn = 3 * 2
iloraz = 3 / 2
reszta = 3 % 2
potęga = 3 ** 2
dzielenie = 3 // 2
```

Niektóre, mimo że *arytmetyczne*, mają zastosowanie też w innych typach.

```python
tekst = tekst + " NAKLEJKA"
print(tekst)
tekst *= 3
print(tekst)
```

Mieszanie typów może jednak mieć różne efekty.

```python

try:
    tekst = tekst + 2
except TypeError:
    print("Nie można dodać str i int. (Normalnie mielibyśmy błąd, tu go przechwytuję)")
```
