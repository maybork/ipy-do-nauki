Python ma narzędzia do kontroli przepływu informacji podobne, przez co intuicyjne, do wielu innych języków. Występują:
- `if`, jeśli warunek jest spełniony wykonuje zagnieżdżony w nim kod
- `elif`, jeśli poprzedni warunek nie został spełniony, sprawdza ten i wykonuje kod, jeśli jest spełniony
- `else` nie przyjmuje żadnego warunku - wykonuje kod, jeśli wszystkie poprzednie warunki zawiodły.
- `while` wykonuje zagnieżdżony kod tak długo, jak jakiś warunek jest prawdziwy

```python
tab = ["jabłko", "gruszka"]

if "boczek" in tab:
    print("W tablicy jest boczek xD")
elif "gruszka" in tab:
    print("olaboga gruszka")
else:
    print("to ja już nie wiem co tam jest")
```

Widać tu keyword `in` - bez zaskoczenia, sprawdza czy dany obiekt jest zawarty w innym.


Bardzo istotnym narzędziem do flow control są funkcje. Nie potrzebujemy definiować, jakiego typu oczekujemy na wejściu lub wyjściu, wystarczy keyword `def` i nazwa.

```python
def hello_world():
    print("no siema")

hello_world()
```

Możemy przyjmować argumenty przy wywołaniu funkcji.

```python
def powitanie(osoba):
    print("hej,", osoba)

powitanie("Maja")
```

Te mogą mieć domyślne wartości, co pozwala nam na tworzenie opcjonalnych argumentów.

```python
def potega(a, n=2):
    return a**n

print(potega(10))
print(potega(10,3))
```

Funkcje na początku sięgają do lokalnych nazw, czyli takich zdefiniowanych przez nie same lub podane jako argument. Dopiero potem sięgają po globalne.

```python
def ssssiema(imie):
    print("hej,", imie)

imie = "amsz"
ssssiema("aszi")
```

Tu, jak widać, funkcja użyła argumentu `imię`, mimo że istniała globalna zmienna o tej samej nazwie. W ten sposób unikamy konfliktów.

Powraca keyword `return`. Kiedy z niego korzystać, a kiedy po prostu printować wynik?
**Moim zdaniem** zawsze `return`. Dlaczego?
- Otrzymany wynik możemy wykorzystać w jakikolwiek sposób (również wyprintować), wyprintowanego nie możemy 
- Taka jest konwencja.

```python
def nium():
    return "siup"

print(nium())
```
