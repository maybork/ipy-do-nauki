Obiekty w Pythonie mogą być *truthy* i *falsy*. To oznacza, że w zależności od ich wartości mają różną wartość logiczną.

```python
liczba = 5

if liczba:
    print("Ta jest truthy, czyli ma logiczną wartość True.")
else:
    print("Nie jest truthy.")
```

```python
liczba = 0

if not liczba:
    print("Ta jest falsy, czyli ma logiczną wartość False.")
else:
    print("Nie jest falsy.")
```

Działa to dla bardzo wielu typów w przewidywalny sposób:
- pusta lista jest `False`
- wartości zerowe są `False`
- `str` bez znaków jest `False`

To pozwala na łatwe sprawdzanie np. obecności elementów w zbiorach, nawet bez korzystania z funkcji `len()`.

```python
lista = []

if lista:
    for el in lista:
        print(el)
```
