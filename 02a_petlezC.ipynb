W iterowaniu po zbiorze może kusić skorzystanie z podobnej do C notacji. W Pythonie możliwe są pętle 'w stylu C' widziane w np. C, C++

```python
array = ["a", "b", "c"]
for i in range(0, len(array)):
    print(array[i])
```

Albo...

```python
i = 0
while i < len(array):
    print(array[i])
    i += 1
```

Ale to jest czytelniejsze i szybsze, a produkuje identyczny wynik!

```python
for element in array:
    print(element)
```

Skakanie po wielu elementach naraz, razem z większymi 'krokami' (co 2, co 3,...) też jest możliwe, bez martwienia się indexami

```python
i = 0
while i < len(array):
    print(array[i])
    i += 2

for element in array[::2]:
    print(element)
```
